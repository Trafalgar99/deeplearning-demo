```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
df=pd.read_excel('酒店数据2.xlsx',index_col=0)
df[:5]
```

（1）读取酒店数据2.xlsx，根据评分和价格信息，绘制散点图。

```python
x=df['评分']
y=df['价格']

plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.title('酒店评分和价格信息散点图',fontsize=20)
plt.xlabel('评分',fontsize=18)
plt.xlabel('价格',fontsize=18)
plt.show()
```

（2）画出油尖旺地区，评分的箱线图。

```python
data=df[df.地区=='油尖旺']['评分']

plt.boxplot(data,whis=2)
plt.title('油尖旺地区酒店评分箱线图',fontsize=20)
plt.show()
```

（3）选出平均价格前5的地区，画出这些地区的评分的箱线图。

```python
data=df['价格'].groupby(df['地区']).mean()
```

```python
data1=df[df.地区=='中西区']['评分']
data2=df[df.地区=='离岛']['评分']
data3=df[df.地区=='沙田']['评分']
data4=df[df.地区=='湾仔']['评分']
data5=df[df.地区=='东区']['评分']

plt.figure(figsize=(10,6))
plt.boxplot([data1,data2,data3,data4,data5],labels=['中西区','离岛','沙田','湾仔','东区'],whis=2)
plt.title('平均价格前五的地区的酒店评分的箱线图',fontsize=20)
plt.show()
```
