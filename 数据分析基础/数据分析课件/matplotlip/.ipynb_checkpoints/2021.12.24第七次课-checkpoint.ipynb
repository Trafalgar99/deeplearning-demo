```python
df=pd.read_excel('movie_data3.xlsx',index_col=0)
df[:5]
```

# 双轴图的画法

```python
import matplotlib.mlab as mlab
from scipy.stats import norm
```

```python
fig=plt.figure(figsize=(10,8))
ax1=fig.add_subplot(111)   #一行一列第一个图
n,bins,patches=ax1.hist(df['评分'],bins=100,color='m')

ax1.set_ylabel('电影数量',fontsize=15)
ax1.set_xlabel('评分',fontsize=15)
ax1.set_title('频率分布直方图',fontsize=20)

y=norm.pdf(bins,df['评分'].mean(),df['评分'].std())
ax2=ax1.twinx()
ax2.plot(bins,y,'b--')
ax2.set_ylabel('概率分布',fontsize=15)
plt.show()
```

根据电影时长和电影评分绘制散点图

```python
x=df['时长']
y=df['评分']

plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.legend()
plt.title('电影时长与评分散点图',fontsize=20)
plt.xlabel('时长',fontsize=18)
plt.ylabel('评分',fontsize=18)
plt.show()
```

```python
#选择原有数据的百分之一
x=df['时长'][::100]
y=df['评分'][::100]

plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.title('电影时长与评分散点图',fontsize=20)
plt.xlabel('时长',fontsize=18)
plt.ylabel('评分',fontsize=18)
plt.show()
```

# 绘制各个地区的评分箱线图

美国电影评分的箱线图

```python
data=df[df.产地=='美国']['评分']

plt.figure(figsize=(10,6))
plt.boxplot(data,whis=2)
plt.title('美国电影评分',fontsize=20)
plt.show()
```

```python
data=df[df.产地=='美国']['评分']

plt.figure(figsize=(10,6))
plt.boxplot(data,whis=2,flierprops={'marker':'o','markerfacecolor':'r','color':'k'})
plt.title('美国电影评分',fontsize=20)
plt.grid()
plt.show()
```

多组数据箱线图

```python
data1=df[df.产地=='中国大陆']['评分']
data2=df[df.产地=='中国香港']['评分']
data3=df[df.产地=='日本']['评分']
data4=df[df.产地=='英国']['评分']
data5=df[df.产地=='法国']['评分']
```

```python
plt.figure(figsize=(12,10))
plt.boxplot([data1,data2,data3,data4,data5],labels=['中国大陆','中国香港','日本','英国','法国'],whis=2)
plt.title('电影评分箱线图',fontsize=20)
plt.grid()
plt.show()
```

vert属性旋转图形

```python
plt.figure(figsize=(12,10))
plt.boxplot([data1,data2,data3,data4,data5],labels=['中国大陆','中国香港','日本','英国','法国'],whis=2,vert=False)
plt.title('电影评分箱线图',fontsize=20)
plt.grid()
plt.show()
```

# 相关系数矩阵图-热力图

```python
data=df[['投票人数','评分','时长']]
data[:5]
```

```python
%pylab inline
result=pd.plotting.scatter_matrix(data[::100],diagonal='kde',color='k',alpha=0.3,figsize=(10,10))
```

```python
%pylab inline
result=pd.plotting.scatter_matrix(data[::100],diagonal='hist',color='k',alpha=0.3,figsize=(10,10))
```

seaborn库画热力图

```python
import seaborn as sns

corr=data.corr()
corr=abs(corr)

fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot()

ax=sns.heatmap(corr,vmax=1,vmin=0,annot=True)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```
