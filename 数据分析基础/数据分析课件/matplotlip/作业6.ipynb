```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

（1）画出$y=x^{2}+2x+1$在区间[-5,3]的函数图像。

```python
x=np.arange(-5,4)
print(x)
y=x*x+2*x+1

plt.plot(x,y)
plt.show()
```

（2）在同一张图中创建两个子图，分别画出sinx和cosx在[-3.14,3.14]上的函数图像。设置线条宽度为2.5.

```python
x=np.arange(-np.pi,np.pi,0.1)

plt.figure(figsize=(10,6))

plt.subplot(211)
plt.plot(x,np.sin(x),linewidth=2.5)

plt.subplot(212)
plt.plot(x,np.cos(x),linewidth=2.5)
```

（3）读取上次作业保存的酒店数据，画出每个地区酒店数量的柱状图，柱状颜色为红色

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
df=pd.read_excel('酒店数据2.xlsx',index_col=0)
df[:5]
```

```python
data=df['地区'].value_counts()
data
```

```python
x=data.index
y=data.values

plt.figure(figsize=(10,6))
plt.bar(x,y,color='g')
plt.show()
```

（4）画出每个价格等级酒店数量的柱状图。

```python
data=df['等级'].value_counts()
data
```

```python
x=data.index
y=data.values

plt.figure(figsize=(10,6))
plt.title('每个价格等级的酒店数量',fontsize=20)
plt.bar(x,y,color='y')
plt.show()
```

（5）画出各个价格等级占比的饼图。

```python
y=data.values

plt.figure(figsize=(6,6))
plt.title('价格等级占比',fontsize=15)

plt.pie(y,labels=data.index,colors='bygr',autopct='%.1f%%')
plt.show()
```
