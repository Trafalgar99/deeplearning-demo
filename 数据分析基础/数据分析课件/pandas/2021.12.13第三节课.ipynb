# 条件选择

选取产地为美国的电影

```python
df['产地']=='美国'
```

```python
df[df['产地']=='美国']
```

```python
df[df['产地']=='美国'][:5]
```

选取产地为美国，且评分大于9的电影

```python
df[(df.产地=='美国')&(df.评分>9)][:5]
```

选取产地为美国或中国，且评分大于9的电影

```python
df[((df.产地=='美国')|(df.产地=='中国大陆'))&(df.评分>9)][:5]
```

# 缺失值和异常值处理

缺失值处理方法

判断缺失值

```python
df.isnull()
```

```python
df[df['名字'].isnull()][:10]
```

```python
df[df['名字'].notnull()][:10]
```

填充缺失值

```python
df[df['投票人数'].isnull()][:10]         #无缺失值
```

```python
df[df['评分'].isnull()][:10] 
```

```python
df['评分'].fillna(np.mean(df['评分']),inplace=True)
```

```python
df[-5:]
```

```python
df.fillna('未知电影')           #所有空值都这样填充
```

删除缺失值

```python
len(df)
```

```python
df2=df.dropna()
```

```python
len(df2)
```

```python
df.dropna(inplace=True)
len(df)
```

处理异常值，一般删除

```python
df[df.投票人数<0]
```

```python
df[df['投票人数']%1!=0]
```

```python
df=df[df.投票人数>0]
df=df[df['投票人数']%1==0]
```
