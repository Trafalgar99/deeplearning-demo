线条属性

```python
x=np.linspace(-np.pi,np.pi)
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.linspace(-np.pi,np.pi)
y=np.sin(x)
plt.plot(x,y,linewidth=4.0,color='r')
plt.show()
```

使用plt.plot()的返回值设置线条属性

```python
plt.plot(x,y,'r-',x,y+1,'g-')
plt.show()
```

```python
line1,line2=plt.plot(x,y,'r-',x,y+1,'g-')
line1.set_antialiased(False)
plt.show()
```

plt.setp()修改线条性质

```python
line=plt.plot(x,y)
```

```python
line=plt.plot(x,y)
plt.setp(line,color='g',linewidth=4)
```

# 子图

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('movie_data3.xlsx',index_col=0)
```

1.绘制每个国家和地区的电影数量的柱状图

```python
data=df['产地'].value_counts()
data
```

```python
x=data.index
y=data.values

plt.figure(figsize=(10,6))
plt.bar(x,y,color='g')
plt.show()
```

```python
x=data.index
y=data.values

plt.figure(figsize=(10,6))
plt.bar(x,y,color='g')
#加入标题
plt.title('各个国家或地区电影数量',fontsize=20)
plt.xlabel('国家或地区',fontsize=18)
plt.ylabel('电影数量',fontsize=18)

plt.show()
```

```python
x=data.index
y=data.values

plt.figure(figsize=(10,6))
plt.bar(x,y,color='g')

plt.title('各个国家或地区电影数量',fontsize=20)
plt.xlabel('国家或地区',fontsize=18)
plt.ylabel('电影数量',fontsize=18)
#调坐标轴
plt.tick_params(labelsize=14)
plt.xticks(rotation=90)
#图上显示数字
for a,b in zip(x,y):
    plt.text(a,b+10,b,ha='center',va='bottom',fontsize=12)
    
plt.show()
```

3.根据电影的长度绘制饼图

```python
#离散化
data=pd.cut(df['时长'],[0,60,90,110,1000]).value_counts()
data
```

```python
y=data.values
y=y/sum(y)

plt.figure(figsize=(6,6))
plt.title('电影时长占比',fontsize=15)

plt.pie(y,labels=data.index,colors='bygr',autopct='%.1f%%')
plt.show()
```

```python
y=data.values
y=y/sum(y)

plt.figure(figsize=(6,6))
plt.title('电影时长占比',fontsize=15)

plt.pie(y,labels=data.index,colors='bygr',autopct='%.1f%%',startangle=90)
plt.legend()
plt.show()
```

```python
y=data.values
#饼图会自动进行归一化
#y=y/sum(y)

plt.figure(figsize=(6,6))
plt.title('电影时长占比',fontsize=15)

patches,l_text,p_text=plt.pie(y,labels=data.index,colors='bygr',autopct='%.1f%%',startangle=90)

for i in p_text:
    i.set_size(15)
    i.set_color('w')
for i in l_text:
    i.set_size(15)
    i.set_color('r')

plt.legend()
plt.show()
```
