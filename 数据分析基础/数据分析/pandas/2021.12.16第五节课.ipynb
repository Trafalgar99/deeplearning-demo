```python
type(group)
```

可以计算分组后的各个统计量

```python
group.sum()
```

计算每年的平均评分

```python
df['评分'].groupby(df['年代']).mean()
```

传入多个分组变量

```python
df.groupby([df['产地'],df['年代']]).mean()
```

获得每个地区每一年电影的评分均值

```python
group=df['评分'].groupby([df['产地'],df['年代']])
```

```python
means=group.mean()
means
```

# 离散化处理

```python
#把评分为9分以上的电影定义为A，7-9定义为B，5-7定义为C，3-5定义为D，小于3定义为E
pd.cut(df['评分'],[0,3,5,7,9,10],labels=['E','D','C','B','A'])
```

```python
df['评分等级']=pd.cut(df['评分'],[0,3,5,7,9,10],labels=['E','D','C','B','A'])
df[:10]
```

根据投票人数来刻画电影的热门

```python
bins=np.percentile(df['投票人数'],[0,20,40,60,80,100])
```

```python
df['热门程度']=pd.cut(df['投票人数'],bins,labels=['E','D','C','B','A'])
df[:10]
```

```python
#筛选
df[(df.评分等级=='E')&(df.热门程度=='A')]
```

```python
df[(df.评分等级=='A')&(df.热门程度=='E')]
```

```python
df.to_excel('movie_data3.xlsx')
```
